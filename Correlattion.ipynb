```python
import yfinance as yf

def get_stock_info(ticker):
    try:
        stock = yf.Ticker(ticker)
        info = stock.info
        return {
            "Sector": info.get("sector", "N/A"),
            "Industry": info.get("industry", "N/A"),
            "Beta": info.get("beta", None)
        }
    except Exception:
        return {"Error": f"Failed to fetch data for {ticker}"}

def find_similar_stocks(main_ticker, candidate_tickers):
    main_info = get_stock_info(main_ticker)
    if "Error" in main_info or main_info["Beta"] is None:
        return f"Error fetching main company data"

    industry, main_beta = main_info["Industry"], main_info["Beta"]
    beta_range = (main_beta * 0.9, main_beta * 1.1)

    similar_stocks = []
    for ticker in candidate_tickers:
        stock_info = get_stock_info(ticker)
        if stock_info.get("Industry") == industry and stock_info.get("Beta") is not None:
            if beta_range[0] <= stock_info["Beta"] <= beta_range[1]:
                similar_stocks.append((ticker, stock_info["Beta"]))

    return sorted(similar_stocks, key=lambda x: abs(x[1] - main_beta))

main_company = "AAPL"
candidate_stocks = [
    "MSFT", "GOOGL", "AMZN", "TSLA", "META", "NVDA", "JPM", "V", "NFLX",
    "PYPL", "INTC", "AMD", "IBM", "BA", "KO", "PEP", "T", "WMT", "DIS",
    "XOM", "CVX", "GS", "C", "BAC", "F", "GM", "GE", "NKE", "MCD",
    "ADBE", "ORCL", "SAP", "CSCO", "CRM", "NOW", "SNOW", "SQ", "UBER", "LYFT",
    "PANW", "ZS", "CRWD", "DOCU", "SHOP", "TWLO", "SNAP", "TTD", "NET", "DDOG"
]

similar_companies = find_similar_stocks(main_company, candidate_stocks)

print(f"Companies in the same industry as {main_company} with similar beta:")
for ticker, beta in similar_companies:
    print(f" - {ticker}: Beta {beta:.2f}")
```
